```python
#Kaan Kılıçarslan 2220674042
#Game Of Life

import tkinter as tk

GRID_SIZE = 20
CELL_SIZE = 20
grid = [[0] * GRID_SIZE for _ in range(GRID_SIZE)]

# Read files and assign them as "read_file"

def read_file(file_name):
    try:
        with open(file_name, 'r') as file:
            for row, line in enumerate(file):
                if row < GRID_SIZE:
                    values = line.replace(',', '').split()  
                    for col in range(min(len(values), GRID_SIZE)):
                        grid[row][col] = int(values[col])
    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
    create_grid()

# Create the grid and 
#visualize it on the canvas

#https://stackoverflow.com/questions/20501512/tkinter-gui-canvas

def create_grid():
    canvas.delete("all")
    
    for row in range(GRID_SIZE):
        for col in range(GRID_SIZE):
            x1, y1 = col * CELL_SIZE, row * CELL_SIZE
            x2, y2 = x1 + CELL_SIZE, y1 + CELL_SIZE
            color = "black" if grid[row][col] else "white"
            canvas.create_rectangle(x1, y1, x2, y2, fill=color)


# Calculate the next generation of the 
#grid based on Game of Life rules in 
#Hacettepe lecture assignment slide

def next_generation():
    new_grid = [[0] * GRID_SIZE for _ in range(GRID_SIZE)]

    for row in range(GRID_SIZE):
        for col in range(GRID_SIZE):
            neighbors = 0
            for i in range(-1, 2):
                for j in range(-1, 2):
                    if i == 0 and j == 0:
                        continue
                    new_row, new_col = row + i, col + j
                    if 0 <= new_row < GRID_SIZE and 0 <= new_col < GRID_SIZE and grid[new_row][new_col] == 1:
                        neighbors += 1
            
            if grid[row][col] == 1:
                if neighbors < 2 or neighbors > 3:
                    new_grid[row][col] = 0
                else:
                    new_grid[row][col] = 1
            elif grid[row][col] == 0:
                if neighbors == 3:
                    new_grid[row][col] = 1
                else:
                    new_grid[row][col] = 0

    for i in range(GRID_SIZE):
        grid[i][:] = new_grid[i][:]

    create_grid()

root = tk.Tk()
root.title("Kaan's Game of Life")

canvas = tk.Canvas(root)
canvas.pack()

next_gen_button = tk.Button(root, text="Next Generation", command=next_generation)
next_gen_button.pack()

#If you get error from file_path, please change the direction of file 

read_file("C:\\Users\\ASUS\\Desktop\\2_Game_of_Life\\glider.txt") 

root.mainloop()
```
